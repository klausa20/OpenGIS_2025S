```python
import pandas as pd
import json 
```

```python
data = json.load(open('lab_2_coffee.json'))
df = pd.DataFrame(data["results"])
```

```python
df['latitude'] = df['geocodes'].apply(lambda x: x['main']['latitude'])
df['longitude'] = df['geocodes'].apply(lambda x: x['main']['longitude'])
```

```python
import folium

center_lat = df['latitude'].mean()
center_long = df['longitude'].mean()
lab_2_map = folium.Map(location = [center_lat, center_long], zoom_start = 17)

for idx, row in df.iterrows():
    folium.Marker(
        location = [row['latitude'], row['longitude']],
        popup = row['name']
    ).add_to(lab_2_map)

lab_2_map
```
